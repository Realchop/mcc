```python
import networkx as nx
from common import *
import matplotlib.pyplot as plt
import iterative
import ilp
import coloring
from pathlib import Path
import pandas as pd
```

```python
def bench(solver, solver_args, path, max_count):
    prep = []
    for file_path in Path(path).iterdir():
        if file_path.is_file():
            df = pd.read_parquet(file_path)
            
            G = nx.from_pandas_edgelist(
                df, 
                source="source", 
                target="target", 
                edge_attr=None
            )
            
            mcn = len(solver(G, *solver_args))
            prep.append({"test" : file_path, "mcn": mcn})
        if max_count and len(prep) == max_count:
            break
    return pd.DataFrame(prep)
```

```python
%%time
df = bench(coloring.minimum_clique_cover, ["greedy"], "../data/nm", 10)
df.describe()
```
