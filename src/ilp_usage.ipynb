```python
import networkx as nx
import pandas as pd
from itertools import product
from ilp import *
```

```python
%%time
ps = ["0.75", "0.5", "0.25"]
names = range(1, 11)

for p, name in product(ps, names):
    test_name = f"{name}_{p}"
    df = pd.read_parquet(f"../data/test/{test_name}.parquet")
    
    G = nx.from_pandas_edgelist(
        df, 
        source="source", 
        target="target", 
        edge_attr=None
    )

    cover = ilp_clique_cover(G)
    with open(f"../data/solutions/{test_name}.txt", "w") as f:
        f.write("\n".join(",".join([str(node) for node in clique]) for clique in cover))
    print(f"Done with: {test_name}")
```
