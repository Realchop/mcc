```python
import networkx as nx
from common import *
import matplotlib.pyplot as plt
import iterative
import ilp
import coloring
from pathlib import Path
import pandas as pd
from itertools import product
```

```python
def bench(solver, solver_args):
    probs = [
        "0.25",
        "0.5",
        "0.75"
    ]
    errors = []
    correct = 0
    total = 0
    for p, n in product(probs, range(1, 11)): 
            solution = []
            with open(f"../data/solutions/{n}_{p}.txt") as f:
                for line in f:
                    solution.append([ int(x) for x in line.split(",")])
                    
            df = pd.read_parquet(f"../data/test/{n}_{p}.parquet")
            
            G = nx.from_pandas_edgelist(
                df, 
                source="source", 
                target="target", 
                edge_attr=None
            )
            
            mcn = solver(G, *solver_args)
            if len(mcn) == len(solution):
                correct += 1
            total += 1
            errors.append(len(mcn) - len(solution))
    
    print(f"{correct}/{total} [{correct/total}]")
    print(f"Error: {sum(errors)}")
    return errors
```

```python
errors = bench(coloring.minimum_clique_cover, ["greedy"])
print(errors)
```

```python
errors = bench(coloring.minimum_clique_cover, ["vote"])
print(errors)
```
